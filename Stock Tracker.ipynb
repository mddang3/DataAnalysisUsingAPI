```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple = yf.Ticker("AAPL")
apple_info = apple.info
apple_share_price_data = apple.history(period='max')
apple_share_price_data.reset_index(inplace=True)

ms = yf.Ticker("MSFT")
ms_info = ms.info
ms_share_price_data = ms.history(period='max')
ms_share_price_data.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize=(8,6))
applplt = plt.plot(apple_share_price_data['Date'], apple_share_price_data['Open'], label='Apple')
msplt = plt.plot(ms_share_price_data['Date'], ms_share_price_data['Open'], label='Microsoft')
fig.legend()
plt.xlabel('Date')
plt.ylabel('Stock Share Price in Dollars')
plt.title('Apple vs Microsoft Stock Price Over Time')
```

```python
apple.dividends
apple.dividends.plot(xlabel='Date', ylabel='Dividends', title='Apple Stock Dividends Over Time')
```

```python
fig2 = plt.figure(figsize=(8,6))
appldivplt = plt.plot(apple_div, label='Apple Dividends')
msdivplt = plt.plot(ms_div, label='Microsoft Dividends')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Dividend Yield')
plt.title('Apple vs Microsoft Dividend Yield')
```
